```python
# By default nltk is not part of Python libraries as in Anaconda. However, we can download the packed as below once.
#import nltk
#nltk.download()          # This will take some time to download

from nltk.tokenize import sent_tokenize     # Library to sentence tokenization

paragraph = "Thanks for taking my courses. You guys rock. Learning is fun. Please take the rest of my courses."
sent_tokenize(paragraph)
```

```python
from nltk.tokenize import word_tokenize

paragraph = "Thanks for taking my courses. You guys rock. Learning is fun. Please take the rest of my courses."
word_tokenize(paragraph)
```
