# Multiple Linear Regression

```python
#Importing pandas library to read CSV data file
import pandas as pd

#Importing numpy to perfom math on arrays
import numpy as np

#Importing statsmodels.formula to fit the regression model
# import statsmodels.formula.api as sm
import statsmodels.regression.linear_model as sm
```

```python
#Reading CSV data file in Python
dataset = pd.read_csv('Companies-spends-and-profits.csv')
```

```python
dataset
```

```python
#Dividing dataset into X and y
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 4].values
```

```python
X
```

```python
y
```

## Encoding categorical data to create the Dummy Variables

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder = LabelEncoder()
X[:, 0] = labelencoder.fit_transform(X[:, 0])
onehotencoder = OneHotEncoder(categorical_features = [0])
X = onehotencoder.fit_transform(X).toarray()
```

```python
X
```

## Adding Intercept column to X

```python
# Creating a new array of given shape (total number of rows in the dataset and add one column) and type, filled with ones.
arr = np.ones((71, 1)).astype(int)

# Now append the Intercept column to X.
# 'axis=1' means we are adding a colomn. 'axis=0' means we are adding a row.
X = np.append(arr, values = X, axis = 1)
```

```python
X
```

```python
# Splitting dataset into Training set and Test set
from sklearn.model_selection import train_test_split
X_train,X_test,y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
# Fitting the first regression model
X_fitted = X_train[:,:]

# Using the Ordinary Least Square (OLS) from 'statsmodels.formula.api' library
reg_1 = sm.OLS(endog = y_train, exog = X_fitted).fit()
reg_1.summary()
```

### Looking at the reg_1 summary, it is found the x2 column has highest 'P-value' (indicating its In-Significant). Hence following  'Backward Elimination' process we remove x2 column and re-run our regression, unti we get columns having 'Significant' 'P-value (which is less than 5% or 0.005)'.

```python
#Fitting the second regression model
X_fitted = X_train[:, [0,1,3,4,5,6,7]]
reg_2 = sm.OLS(endog = y_train, exog = X_fitted).fit()
reg_2.summary()
```

```python
#Fitting the second regression model
X_fitted = X_train[:, [0,3,4,5,6,7]]
reg_3 = sm.OLS(endog = y_train, exog = X_fitted).fit()
reg_3.summary()
```

```python
#Fitting the second regression model
X_fitted = X_train[:, [0,3,4,5,7]]
reg_4 = sm.OLS(endog = y_train, exog = X_fitted).fit()
reg_4.summary()
```

```python
#Fitting the second regression model
X_fitted = X_train[:, [0,3,4,5]]
reg_5 = sm.OLS(endog = y_train, exog = X_fitted).fit()
reg_5.summary()
```

```python
#Fitting the second regression model
X_fitted = X_train[:, [0,3,5]]
reg_6 = sm.OLS(endog = y_train, exog = X_fitted).fit()
reg_6.summary()
```

```python
#Fitting the second regression model
X_fitted = X_train[:, [0,5]]
reg_7 = sm.OLS(endog = y_train, exog = X_fitted).fit()
reg_7.summary()
```

### Looking at the above summary, its found that 'x1' which corresponds to IT column in our main dataset. So 'IT' has the high power to predict the profits. In other words, the 'IT' expenses can reflect on the profit. So higher expense on 'IT', higher the profit the company can get.

### Additionally, 'R-squared' and 'Adj. R-squared' is about 30%, which means this model is not really powerful. It shows only 30% of the change in the dependent variable.

### Finally, lets try to test this model.

```python
X_for_test = X_test[:, [0,5]]

y_prdicted = reg_7.predict(X_for_test)
```
