```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
master_frame_venture = pd.read_csv("master_df_CountryAnalysis.csv", encoding='ISO-8859-1')
```

```python
master_frame_venture.head(50)
```

```python
master_frame_venture['category_list']
```

```python
master_frame_venture['category_list'] = master_frame_venture['category_list'].str.split('|').str[0]
```

```python
master_frame_venture.head(50)
```

```python
mapping_frame = pd.read_csv("mapping.csv", encoding='ISO-8859-1')
```

```python
mapping_frame.info()
```

```python
mapping_frame.head()
```

```python
# Dropping rows whose 'category_list' column is NaN.
mapping_frame = mapping_frame.dropna(subset=['category_list'])
```

```python
# Coverting the mapping column 'category_list' in both df to lowercase.
master_frame_venture['category_list'] = master_frame_venture['category_list'].str.lower()
mapping_frame['category_list'] = mapping_frame['category_list'].str.lower()
```

```python
mapping_frame.head(50)
```

```python
# Replacing '0' with 'na' under 'category_list'.
mapping_frame['category_list'] = mapping_frame['category_list'].apply(lambda x: x.replace('0', 'na'))
mapping_frame.head(50)
```

```python
# 'category_list' items present in mapping_frame, but not in master_frame_venture
mapping_frame.loc[~mapping_frame['category_list'].isin(master_frame_venture['category_list'])]
```

```python
# 'category_list' items present in master_frame_venture, but not in mapping_frame
master_frame_venture.loc[~master_frame_venture['category_list'].isin(mapping_frame['category_list'])]
```

```python
# Merging mapping_frame df into master_frame_venture df
master_frame_venture = pd.merge(master_frame_venture, mapping_frame, how='inner', left_on='category_list', right_on='category_list')
```

```python
master_frame_venture.info()
```

```python
master_frame_venture.head()
```

#### Now we have a total of 18 columns post merger. However, the newly merged 10 columns are nothing but one's and zero's, all indicating one only value for the given category. It would be convenient to do further analysis if we could shrink down these 10 columns into one column - say 'sector'.

#### We will use pandas 'melt' function to acheive this

```python
# Store the newly added 10 columns into a seperate Series.
value_vars = master_frame_venture.columns[9:18]
print(value_vars)

# Take the remaining columns as seperate Series.
id_vars = master_frame_venture.columns.difference(value_vars)
# id_vars = np.setdiffid(df.columns, value_vars)                 # Alternative ways of doing the same job.
print(id_vars)
```

```python
# Calling the 'melt' function
master_frame_venture = pd.melt(master_frame_venture, id_vars=list(id_vars), value_vars=list(value_vars))
```

```python
master_frame_venture
```

```python
# Deleting 'value' column rows having 0, which is not required.
master_frame_venture.drop(master_frame_venture[master_frame_venture['value'] == 0].index, axis=0, inplace=True)
```

```python
# Cross checking
len(master_frame_venture[master_frame_venture['value'] == 0])
```

```python
master_frame_venture.shape
```

#### Got back the same row numbers as before post deleting un wanted 'value = 0' rows.

```python
# Now, drop the 'value' column and rename 'variable' column to may be 'sector'.
master_frame_venture.drop(['value'], axis=1, inplace=True)

master_frame_venture.rename(columns={'variable': 'sector'}, inplace=True)
```

```python
# Saving the merged master frame for further analysis
master_frame_venture.to_csv("master_df_SectorAnalysis1.csv", sep=',', index=False)
```
