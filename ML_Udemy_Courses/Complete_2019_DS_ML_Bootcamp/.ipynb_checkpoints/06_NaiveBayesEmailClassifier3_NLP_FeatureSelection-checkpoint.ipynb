### Word Stems and Stemming

```python
sample_text = "All work and no play makes jack a dull boy. To be or not to be. \
               Nobody expects the Spanish Inquisition!"
sample_words = word_tokenize(sample_text.lower())

#stemmer = PorterStemmer() # PorterStemmer package supports 'English' language.
stemmer = SnowballStemmer('english') # We could use SnowballStemmer package, which supports multiple languages.

filtered_sample_words = []
# Challenge: append non-stop words to filtered_sample_words
for word in sample_words:
    if word not in stop_words:
        stemmed_word = stemmer.stem(word)
        filtered_sample_words.append(stemmed_word)
        
print(filtered_sample_words)
```

In the above output, one can notice that the word 'makes' stems to 'make', 'nobody' -> 'nobodi', 'expects' -> 'expect', etc

### Functions for Email Processing

```python
def clean_message(message, stemmer=PorterStemmer(), 
                  stop_words=set(stopwords.words('english'))):
    
    # Converts to Lower Case and splits up the words.
    words = word_tokenize(message.lower())
    
    filtered_words = []
    
    for word in words:
        # Remove stop words and punctuations.
        if word not in stop_words and word.isalpha():
            filtered_words.append(stemmer.stem(word))
    
    return filtered_words
```

Lets load a sample email to test the above function.

```python
fileStream = open(EXAMPLE_FILE, encoding='latin-1')

is_body = False
lines = []

for line in fileStream:
    if is_body:
        lines.append(line)
    elif line == '\n':
        is_body = True
        
fileStream.close()

email_body = '\n'.join(lines)
print(email_body)
```

```python
clean_message(email_body)
```

```python
# Challenge: Modify the 'clean_message' function to remove HTML Tags as well.
def clean_message_no_html(message, stemmer=PorterStemmer(), 
                  stop_words=set(stopwords.words('english'))):
    
    #Removing HTML tags.
    soup = BeautifulSoup(message, 'html.parser')
    cleaned_message = soup.get_text()
    
    # Converts to Lower Case and splits up the words.
    words = word_tokenize(cleaned_message.lower())
    
    filtered_words = []
    
    for word in words:
        # Remove stop words and punctuations.
        if word not in stop_words and word.isalpha():
            filtered_words.append(stemmer.stem(word))
    
    return filtered_words
```

```python
%%time

# Challenge 1: Use python list comprehension and 
#              then find the total number of words in our cleaned dataset of spam email bodies.

cleaned_spam_emails = data[data.CATAGORY == CAT_SPAM].MESSAGE.apply(clean_message_no_html)

flat_spam_emails_list = [item for sublist in cleaned_spam_emails for item in sublist]

# Dump the list onto a Panda Series
spam_email_words = pd.Series(flat_spam_emails_list).value_counts() # Fetches the unique words count

spam_email_words.shape[0] # The total number of unique words in our cleaned dataset of spam email
```

```python
%%time

# Challenge 2: Find the total number of words in normal emails in the dataset.

cleaned_ham_emails = data[data.CATAGORY == CAT_HAM].MESSAGE.apply(clean_message_no_html)

flat_ham_emails_list = [item for sublist in cleaned_ham_emails for item in sublist]

# Dump the list onto a Panda Series
ham_email_words = pd.Series(flat_ham_emails_list).value_counts() # Fetches the unique words count

ham_email_words.shape[0] # The total number of unique words in our cleaned dataset of spam email
```

### Generate Vocabolary and Dictionary

```python
stemmed_nested_list = data.MESSAGE.apply(clean_message_no_html)
flat_stemmed_list = [item for sublist in stemmed_nested_list for item in sublist]
```

```python
unique_words = pd.Series(flat_stemmed_list).value_counts()
print('Number of unique words', unique_words.shape[0])
unique_words.head()
```

```python
# Challenge: Create subset of the series called 'frequent_words'that only contains the most common 2,500 words out of the total.
# Print out the top 10 words.

frequent_words = unique_words[:VOCABULARY_SIZE]
print('Most common words: \n', frequent_words[:10])
```

### Create Vocabulary DataFrame with a WORD_ID

```python
# To make a Vocabulary DataFrame we need to convert items with in 'frequest_words' which is a pd.Series into and pd.DataFrame.
word_ids = list(range(0, VOCABULARY_SIZE))
vocab = pd.DataFrame({'VOCAB_WORD' : frequent_words.index.values}, index=word_ids)
vocab.index.name = 'WORD_ID'
vocab.head()
```

### Save the Vocabulary as a CSV File

```python
vocab.to_csv(WORD_ID_FILE, index_label=vocab.index.name, header=vocab.VOCAB_WORD.name)
```
