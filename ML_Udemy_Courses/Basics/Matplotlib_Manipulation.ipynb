# matplotlib
matplotlib is simply a library for data visualization.
The three most common graphs you'll see in matplotlib are: line, scatter and histogram

```python
import matplotlib.pyplot as plt # Use pyplot to plot some data to the canvas in memory
import numpy as np
```

```python
x = np.linspace(0 ,10, 100)
```

```python
plt.plot(x, x, label='linear')
plt.legend()
plt.show()
```

```python
# pylab is module in matplotlib that get's installed alongside matplotlib
import numpy as np
import pylab as pl

ds = np.random.normal(5.0, 3.0, 1000)

pl.hist(ds)
pl.xlabel('ds')
pl.show()
```
