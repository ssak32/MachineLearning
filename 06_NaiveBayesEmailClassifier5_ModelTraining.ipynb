### Creating Full Matrix

```python
def make_full_matrix(sparse_matrix, nr_words, doc_idx=0, word_idx=1, cat_idx=2, freq_idx=3):
    
    column_names = ['DOC_ID', 'CATEGORY'] + list(range(0, nr_words))
    index_names = np.unique(sparse_matrix[:, 0])
    
    full_matrix_df = pd.DataFrame(columns=column_names, index=index_names)
    full_matrix_df.fillna(value=0, inplace=True)

    for i in range(0, sparse_matrix.shape[0]):
        doc_nr = sparse_matrix[i][doc_idx]
        word_id = sparse_matrix[i][word_idx]
        label = sparse_matrix[i][cat_idx]
        occurrence = sparse_matrix[i][freq_idx]
        
        full_matrix_df.at[doc_nr, 'DOC_ID'] = doc_nr
        full_matrix_df.at[doc_nr, 'CATEGORY'] = label
        full_matrix_df.at[doc_nr, word_id] = occurrence
        
        
    full_matrix_df.set_index('DOC_ID', inplace=True)
    
    return full_matrix_df
    
```

```python
%%time
full_train_matrix = make_full_matrix(sparse_train_data, VOCABULARY_SIZE)
full_train_matrix.head()
```

## Training the Naive Bayes Model
### Calculating the Probability of Spam

```python
# Challenge: Calculate the probability of spam - the percent of spam messages in the training dataset.
# Store this value in a variable called prob_spam.

print('Total number of emails:', len(full_train_matrix))
print('Total number of spam emails:', len(full_train_matrix[full_train_matrix.CATEGORY == 1]))

prob_spam = (len(full_train_matrix[full_train_matrix.CATEGORY == 1]) / len(full_train_matrix)) * 100
print('Percentage of spam emails:', prob_spam)
```

### Number of Tokens in Spam and Ham Emails

**Challenge**
1. Create a subset named 'spam_lengths', then count the total number of words that occur in spam emails
2. Also, create similar subset named 'ham_lengths', then count the total number of words that occur in ham emails.

```python
# Calculating spam lengths.
spam_lengths = full_train_matrix[full_train_matrix.CATEGORY == 1]
spam_lengths.drop(columns=['CATEGORY'], inplace=True)
spam_wc = spam_lengths.sum().sum()
print('The total number of words in Spam emails are:', spam_wc)
```

### P(Token|Spam) - Probability that a Token Occurs given the Email is Spam

```python
prob_tokens_spam = summed_spam_tokens / (spam_wc + VOCABULARY_SIZE)
prob_tokens_spam[:5]
```

```python
prob_tokens_spam.sum()
```

### P(Token|Ham) - Probability that a Token Occurs given the Email is Ham

```python
prob_tokens_ham = summed_ham_tokens / (ham_wc + VOCABULARY_SIZE)
prob_tokens_ham[:5]
```

```python
prob_tokens_ham.sum()
```

### P(Token) -  Probability of Token Occurs

```python
prob_tokens_all = full_train_features.sum(axis=0) / (total_wc)
prob_tokens_all.sum()
```
